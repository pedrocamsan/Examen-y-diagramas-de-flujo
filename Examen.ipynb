```
a = float(input("Dame el valor de x al cuadrado: "))
b = float(input("Dame el valor de x lineal: "))
c = float(input("Dame el valor de el termino independiente: "))

dis= (b**2)-4*a*c

while (dis<0):
  print("Error el discriminante tiene que ser positivo")
  a = float(input("Dame el valor de x al cuadrado: "))
  b = float(input("Dame el valor de x lineal: "))
  c = float(input("Dame el valor del termino independiente: "))
  dis= (b**2)-4*a*c
     
if dis==0:
    x= -b/(2*a)
    print("X solo tiene una raiz = ",x)
else: 
    x1= (-b+(dis**0.5))/(2*a)
    x2= (-b-(dis**0.5))/(2*a)
    print("La primer raiz de X1 =",x1)
    print("La segunda raiz de X2 =",x2)
```
